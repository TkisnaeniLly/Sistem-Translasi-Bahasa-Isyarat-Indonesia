```python
import os
import shutil

source_dir = 'datasetBISINDO'  # atau dataset_raw
target_dir = 'dataset_clean'

# cek dulu apakah folder sumber ada
if not os.path.exists(source_dir):
    raise FileNotFoundError(f"Folder sumber '{source_dir}' tidak ditemukan! Pastikan namanya benar.")

if not os.path.exists(target_dir):
    os.makedirs(target_dir)

for file in os.listdir(source_dir):
    if file.endswith('.jpg') or file.endswith('.png'):
        label = file.split('_')[0].upper()
        label_dir = os.path.join(target_dir, label)
        os.makedirs(label_dir, exist_ok=True)
        shutil.copy(os.path.join(source_dir, file), label_dir)
```
