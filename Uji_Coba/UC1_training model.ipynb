```python
# ==============================
# üåü Tahap 4: Training Model BISINDO
# ==============================

# 1Ô∏è‚É£ Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# 2Ô∏è‚É£ Baca dataset hasil ekstraksi tahap 3
data_path = 'dataset_ready/hand_landmarks.csv'
data = pd.read_csv(data_path)

print("Jumlah data:", len(data))
print("Kolom dataset:", list(data.columns))
data.head()

# 3Ô∏è‚É£ Pisahkan fitur (X) dan label (y)
X = data.drop('label', axis=1)
y = data['label']

# 4Ô∏è‚É£ Encode label huruf menjadi angka
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# 5Ô∏è‚É£ Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Jumlah data latih:", len(X_train))
print("Jumlah data uji:", len(X_test))

# 6Ô∏è‚É£ Buat model (Random Forest)
model = RandomForestClassifier(
    n_estimators=200, 
    random_state=42,
    n_jobs=-1
)

# 7Ô∏è‚É£ Latih model
print("üîÅ Sedang melatih model... tunggu sebentar.")
model.fit(X_train, y_train)
print("‚úÖ Pelatihan selesai!")

# 8Ô∏è‚É£ Evaluasi model
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\nüéØ Akurasi Model: {acc * 100:.2f}%\n")
print("üìä Laporan Klasifikasi:")
print(classification_report(y_test, y_pred))

# 9Ô∏è‚É£ Simpan model dan label encoder untuk tahap berikutnya
joblib.dump(model, 'bisindo_model.pkl')
joblib.dump(encoder, 'label_encoder.pkl')

print("\nüíæ Model dan encoder berhasil disimpan!")
print("   - Model: bisindo_model.pkl")
print("   - Encoder: label_encoder.pkl")
```
